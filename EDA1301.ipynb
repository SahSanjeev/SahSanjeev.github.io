```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import Legend

def make_dashboard(x, gdp_change, unemployment, title, file_name, legend=None):
    output_notebook()

    # Create a new plot with a title and axis labels
    p = figure(title=title, x_axis_label='Year', y_axis_label='%')

    # Line glyphs
    p.line(x, gdp_change, line_color="blue", line_width=4, legend_label="GDP Change")
    p.line(x, unemployment, line_color="red", line_width=4, legend_label="Unemployment")

    # Add legend if provided
    if legend:
        p.legend.title = legend
        p.legend.label_text_font_size = '10pt'
        p.legend.title_text_font_size = '12pt'

    # Show the plot
    show(p)

# Example usage
x = [1, 2, 3, 4, 5]
gdp_change = [0.1, 0.2, -0.1, -0.2, 0.15]
unemployment = [5, 4, 5, 3, 6]
title = "Economic Data"
file_name = "dashboard.html"
legend = "My Legend"

make_dashboard(x, gdp_change, unemployment, title, file_name, legend=legend)
```
